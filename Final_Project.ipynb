```python
import ast
rating_names = set()
for index, row in ted_clean.iterrows():
    ratings = ast.literal_eval(row['ratings'])
    for rating in ratings:
        rating_names.add(rating['name'])
print(rating_names)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
rating_names2 = str(rating_names)
```

```python
wordcloud = WordCloud(background_color='black',
                      max_words= 14,
                      colormap='Blues').generate(rating_names2)
fig = plt.figure(1,figsize=(12,10))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis('off')
plt.show()
```

# Names of Ratings that are Assigned to Each Talk



```python
from collections import defaultdict
rating_info = defaultdict(list)
for index, row in ted_clean.iterrows():
    rating = ast.literal_eval(row['ratings'])
    rating_info['id'].append(row['id'])
    names = set()
    for rating in rating:
        rating_info[rating['name']].append(rating['count'])
        names.add(rating['name'])

rating_info = pd.DataFrame(rating_info)

rating_info.head()
```

```python
ted_talks = pd.concat([rating_info, ted_clean], axis = 1)
```

```python
ted_talks.head()
```

```python
duration_min = (ted_clean['duration'])/60
```

```python
ted_talks['duration_min'] = duration_min
```

```python
ted_talks['tags']
```

```python
tag_names = defaultdict(list)
for index, row in ted_clean.iterrows():
    tags_list = ast.literal_eval(row['tags'])
    for item in tags_list: 
        tag_names['id'].append(row['id'])
        tag_names['tags'].append(item)
```

```python
tag_names['tags']
```

```python
tags = set(tag_names['tags'])
```

```python
tags = str(tags)
```

```python
wordcloud = WordCloud(background_color='black',
                      max_words= 100,
                      colormap='Blues').generate(tags)
fig = plt.figure(1,figsize=(12,10))
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis('off')
plt.show()
```
